```python
# Parameters for the ellipse (convex set)
a = 2
b = 1

# Creating an ellipse
theta = np.linspace(0, 2*np.pi, 100)
x_ellipse = a * np.cos(theta)
y_ellipse = b * np.sin(theta)

# Parameters for the "L" shape (non-convex set)
# Defining the "L" shape using line segments
x_L = [-1, -1, 1, 1, 0, 0, -1]
y_L = [-2, 1, 1, 0, 0, -2, -2]

# Plotting the convex and non-convex sets
plt.figure(figsize=(10, 3), dpi=300)

# Plotting the convex set (ellipse)
plt.subplot(1, 2, 1)
plt.plot(x_ellipse, y_ellipse, color=edge_color)
plt.plot([-1, 1], [0, 0], color=edge_color)  # Line inside the ellipse
plt.scatter([-1, 1], [0, 0], color=point_color)
plt.title('Convex Set (Ellipse)')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axis('off')

# Plotting the non-convex set ("L" shape)
plt.subplot(1, 2, 2)
plt.plot(x_L, y_L, color=edge_color)
plt.plot([-0.5, 0.75], [-1, 0.5], color=edge_color)  # Line inside the ellipse
plt.scatter([-0.5, 0.75], [-1, 0.5], color=point_color)
plt.title('Non-Convex Set (L Shape)')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axis("off")
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# file_path = "../images/convex_set.png"
# plt.savefig(file_path, dpi=300)

plt.tight_layout()
plt.show()
```

```python
p1 = gnbu_colormap(0.2)
p2 = gnbu_colormap(0.4)
p3 = gnbu_colormap(0.6)
p4 = gnbu_colormap(0.8)
```

```python
# Define the equations for the lines
x_values = np.linspace(-2, 3, 400)
y1_values = x_values - 1
y2_values = -x_values + 2

# Create the plot again with the intersected region in a different color
plt.figure(figsize=(4, 4), dpi=300)

# Fill the half-planes but not the intersection
plt.fill_between(
    x_values,
    y1_values,
    3,
    where=(x_values <= 3),
    color=p1,
    alpha=0.5,
    label="y >= x - 1",
)
plt.fill_between(
    x_values,
    -3,
    y2_values,
    where=(x_values >= -2),
    color=p4,
    alpha=0.5,
    label="y <= -x + 2",
)

# Fill the intersection region with a different color
plt.fill_between(
    x_values,
    y1_values,
    y2_values,
    where=(y1_values <= y2_values),
    color=p3,
    alpha=0.7,
    label="Intersection",
)

# Plot the lines defining the half-planes
plt.plot(x_values, y1_values, color=p1, label="y = x - 1")
plt.plot(x_values, y2_values, color=p4, label="y = -x + 2")

# Adjust plot settings
plt.xlim(-2, 3)
plt.ylim(-3, 3)
plt.legend()
plt.axis('off')
# plt.title("Intersection of Two Half-Planes")
plt.tight_layout()

plt.show()
```

```python
# Define colors from the GnBu colormap
gnbu_colormap = plt.colormaps['GnBu']
edge_color = gnbu_colormap(0.7)
point_color = gnbu_colormap(0.5)
voronoi_color = gnbu_colormap(0.4)  # point
circle_color = gnbu_colormap(0.2)  # point
```

```python
# Define the equations for the lines
x_values = np.linspace(-2, 3, 400)
y1_values = x_values - 1
y2_values = -x_values + 2

# Create the plot again with the intersected region in a different color
plt.figure(figsize=(4, 4), dpi=300)

# Fill the half-planes but not the intersection
plt.fill_between(
    x_values,
    y1_values,
    3,
    where=(x_values <= 3),
    color=p1,
    alpha=0.5,
    label="y >= x - 1",
)
plt.fill_between(
    x_values,
    -3,
    y2_values,
    where=(x_values >= -2),
    color=p4,
    alpha=0.5,
    label="y <= -x + 2",
)

# Fill the intersection region with a different color
plt.fill_between(
    x_values,
    y1_values,
    y2_values,
    where=(y1_values <= y2_values),
    color=p3,
    alpha=0.7,
    label="Intersection",
)

# Plot the lines defining the half-planes
plt.plot(x_values, y1_values, color=p1, label="y = x - 1")
plt.plot(x_values, y2_values, color=p4, label="y = -x + 2")

# Adjust plot settings
plt.xlim(-2, 3)
plt.ylim(-3, 3)
plt.legend()
plt.axis('off')
# plt.title("Intersection of Two Half-Planes")
plt.tight_layout()

plt.show()

from scipy.spatial import ConvexHull

# Random points in 2-D
# np.random.seed(0)  # For reproducibility
num_points = 10
radius = 50
points = radius * np.random.random((num_points, 2))

# Compute the Convex Hull
hull = ConvexHull(points)

# Create the plot with the filled convex hull polygon
plt.figure(figsize=(4,4), dpi=300)

# Draw the convex hull as a filled polygon
polygon = plt.Polygon(
    points[hull.vertices],
    closed=True,
    fill=True,
    edgecolor=edge_color,
    facecolor=voronoi_color,
    alpha=0.4
)
plt.gca().add_patch(polygon)

# Draw the points
plt.plot(points[:, 0], points[:, 1], "o", markersize=8, color=point_color)

# Configure plot
# plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.gca().set_aspect("equal", adjustable="box")
plt.axis("off")  # Hide axes
plt.tight_layout()
plt.show()
```

```python
# Define colors for the half-plane plot
p1 = gnbu_colormap(0.7)
p3 = gnbu_colormap(0.3)
p4 = gnbu_colormap(0.5)

# Define the equations for the lines for the half-plane plot
x_values = np.linspace(-2, 3, 400)
y1_values = x_values - 1
y2_values = -x_values + 2

# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(8, 4), dpi=300)

# Half-planes plot
# Fill the half-planes but not the intersection
axs[0].fill_between(
    x_values,
    y1_values,
    3,
    where=(x_values <= 3),
    color=p1,
    alpha=0.5,
    label="y >= x - 1",
)
axs[0].fill_between(
    x_values,
    -3,
    y2_values,
    where=(x_values >= -2),
    color=p4,
    alpha=0.5,
    label="y <= -x + 2",
)

# Fill the intersection region with a different color
axs[0].fill_between(
    x_values,
    y1_values,
    y2_values,
    where=(y1_values <= y2_values),
    color=p3,
    alpha=0.7,
    label="Intersection",
)

# Plot the lines defining the half-planes
axs[0].plot(x_values, y1_values, color=p1, label="y = x - 1")
axs[0].plot(x_values, y2_values, color=p4, label="y = -x + 2")

# Adjust plot settings for half-planes plot
axs[0].set_xlim(-2, 3)
axs[0].set_ylim(-3, 3)
axs[0].legend()
axs[0].axis('off')

# Convex hull plot
# Compute the Convex Hull for the second plot
num_points = 10
radius = 50
np.random.seed(125)  # Use the same seed for reproducibility
points = radius * np.random.random((num_points, 2))
hull = ConvexHull(points)

# Draw the convex hull as a filled polygon
polygon = plt.Polygon(
    points[hull.vertices],
    closed=True,
    fill=True,
    edgecolor=edge_color,
    facecolor=voronoi_color,
    alpha=0.4
)
axs[1].add_patch(polygon)

# Draw the points
axs[1].plot(points[:, 0], points[:, 1], "o", markersize=8, color=point_color)

# Adjust plot settings for convex hull plot
axs[1].set_aspect("equal", adjustable="box")
axs[1].axis("off")  # Hide axes

plt.tight_layout()
plt.show()
```
