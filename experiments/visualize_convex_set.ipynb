```python
# Define colors from the GnBu colormap
gnbu_colormap = plt.colormaps['GnBu']
edge_color = gnbu_colormap(0.7)
point_color = gnbu_colormap(0.5)  # point
```

```python
# Parameters for the ellipse (convex set)
a = 2
b = 1

# Creating an ellipse
theta = np.linspace(0, 2*np.pi, 100)
x_ellipse = a * np.cos(theta)
y_ellipse = b * np.sin(theta)

# Parameters for the "L" shape (non-convex set)
# Defining the "L" shape using line segments
x_L = [-1, -1, 1, 1, 0, 0, -1]
y_L = [-2, 1, 1, 0, 0, -2, -2]

# Plotting the convex and non-convex sets
plt.figure(figsize=(10, 3), dpi=300)

# Plotting the convex set (ellipse)
plt.subplot(1, 2, 1)
plt.plot(x_ellipse, y_ellipse, color=edge_color)
plt.plot([-1, 1], [0, 0], color=edge_color)  # Line inside the ellipse
plt.scatter([-1, 1], [0, 0], color=point_color)
plt.title('Convex Set (Ellipse)')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axis('off')

# Plotting the non-convex set ("L" shape)
plt.subplot(1, 2, 2)
plt.plot(x_L, y_L, color=edge_color)
plt.plot([-0.5, 0.75], [-1, 0.5], color=edge_color)  # Line inside the ellipse
plt.scatter([-0.5, 0.75], [-1, 0.5], color=point_color)
plt.title('Non-Convex Set (L Shape)')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axis("off")
# plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# file_path = "../images/convex_set.png"
# plt.savefig(file_path, dpi=300)

plt.tight_layout()
plt.show()
```

```python
p1 = gnbu_colormap(0.2)
p2 = gnbu_colormap(0.4)
p3 = gnbu_colormap(0.6)
p4 = gnbu_colormap(0.8)
```

```python
# Define the equations for the lines
x_values = np.linspace(-2, 3, 400)
y1_values = x_values - 1
y2_values = -x_values + 2

# Create the plot again with the intersected region in a different color
plt.figure(figsize=(4, 4), dpi=300)

# Fill the half-planes but not the intersection
plt.fill_between(
    x_values,
    y1_values,
    3,
    where=(x_values <= 3),
    color=p1,
    alpha=0.5,
    label="y >= x - 1",
)
plt.fill_between(
    x_values,
    -3,
    y2_values,
    where=(x_values >= -2),
    color=p4,
    alpha=0.5,
    label="y <= -x + 2",
)

# Fill the intersection region with a different color
plt.fill_between(
    x_values,
    y1_values,
    y2_values,
    where=(y1_values <= y2_values),
    color=p3,
    alpha=0.7,
    label="Intersection",
)

# Plot the lines defining the half-planes
plt.plot(x_values, y1_values, color=p1, label="y = x - 1")
plt.plot(x_values, y2_values, color=p4, label="y = -x + 2")

# Adjust plot settings
plt.xlim(-2, 3)
plt.ylim(-3, 3)
plt.legend()
plt.axis('off')
# plt.title("Intersection of Two Half-Planes")
plt.tight_layout()

plt.show()
```
