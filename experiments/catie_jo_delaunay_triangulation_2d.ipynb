```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
from time import sleep
```

```python
points = np.array([[-500, -500], [500, 500], [500, -500], [-500, 500]])
```

So according to the problem, you start off with a blank square cut into two triangles, as shown in the image above. Now what? For each point, you need to do three steps:

1. Insert the point onto the graph
2. Draw connecting lines to the vertices of the enclosing triangle
3. Flip illegal edges (aka maximize the minimum angle)

Any two adjacent triangles form a kite/diamond shape. If the kite is convex, look at the shared edge. That edge is illegal if the sum of the angles of the non-connected vertices is greater than 180 degrees. In other words, you want to maximize the minimum angle in these two triangles.



```python
def plotter(p):
  tri = Delaunay(p)
  plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
  plt.plot(points[:,0], points[:,1], 'o')
  # plt.draw()
```

```python
points_to_add = [
    [-150, -100],
    [150, 100],
    [150, -150],
    [150, 150],
    [0, 300],
    [300, 100],
    [0, 50],
    [100, -20],
    [-100, -350],
    [-200, 50],
]
```

```python
for point in points_to_add:
    points = np.append(points, [point], axis=0)
    plotter(points)
    plt.pause(1)
    plt.draw()
```

```python
points = np.append(points, [[-150, -100]], axis=0)
plotter(points)
```
