```python
# Define the radius of the circle and the circle center
radius = 1
center_x = 0
center_y = 0

# Define the angles in radians for each point on the circle at 45 degree intervals starting from 45 degrees
angles = np.radians([45, 45 + 90, 45 + 180, 45 + 270])

# Calculate the x and y coordinates for each point on the circle
points = np.array([[radius * np.cos(angle), radius * np.sin(angle)] for angle in angles])

# Calculate the circle's points for plotting
theta = np.linspace(0, 2 * np.pi, 100)
circle_x = center_x + radius * np.cos(theta)
circle_y = center_y + radius * np.sin(theta)

# Compute the Delaunay triangulation and Voronoi diagram for the points
tri = Delaunay(points)
vor = Voronoi(points)
```

```python
# Define the angles in radians for three points on the circle at 90 degrees and 120 degrees apart
angles_three_points = np.radians([90, 90 + 120, 90 + 240])

# Calculate the x and y coordinates for the three points on the circle
points_three = np.array(
    [
        [radius * np.cos(angle), radius * np.sin(angle)]
        for angle in angles_three_points
    ]
)

# Compute the Delaunay triangulation and Voronoi diagram for the three points
tri_three = Delaunay(points_three)
vor_three = Voronoi(points_three)

# Define the radius of the circle and the circle center
radius = 1
center_x = 0
center_y = 0

# Calculate the circle's points for plotting
theta = np.linspace(0, 2 * np.pi, 100)
circle_x = center_x + radius * np.cos(theta)
circle_y = center_y + radius * np.sin(theta)
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(11, 3), dpi=300)

# First subplot for four points
# Plot the Voronoi diagram
voronoi_plot_2d(
    vor,
    show_vertices=False,
    line_colors=voronoi_color,
    line_width=2,
    line_alpha=0.6,
    show_points=False,
    ax=axs[0],
)
# Plot the Delaunay triangulation
axs[0].triplot(
    points[:, 0], points[:, 1], tri.simplices, color=edge_color, lw=2
)
# Plot the points
axs[0].plot(points[:, 0], points[:, 1], "o", color=point_color, label="Site")
# Plot the circle
axs[0].plot(
    circle_x, circle_y, label="Circle", color=circle_color, linestyle="dotted"
)
axs[0].plot([0], [0], "o", color=edge_color, label="Common point")

# Second subplot for three points
# Plot the Voronoi diagram
voronoi_plot_2d(
    vor_three,
    show_vertices=False,
    line_colors=voronoi_color,
    line_width=2,
    line_alpha=0.6,
    show_points=False,
    ax=axs[1],
)
# Plot the Delaunay triangulation
axs[1].triplot(
    points_three[:, 0],
    points_three[:, 1],
    tri_three.simplices,
    color=edge_color,
    lw=2,
)
# Plot the points
axs[1].plot(points_three[:, 0], points_three[:, 1], "o", color=point_color)
# Plot the circle
axs[1].plot(circle_x, circle_y, color=circle_color, linestyle="dotted")
axs[1].plot([0], [0], "o", color=edge_color)
axs[1].plot(
    [],
    [],
    color=voronoi_color,
    linewidth=2,
    alpha=0.6,
    label="Voronoi Region",
    linestyle="--",
)

for ax in axs:
    ax.set_xlim([-1.2, 1.2])
    ax.set_ylim([-1.2, 1.2])
    ax.set_aspect("equal")
    ax.axis("off")

fig.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay

# Define the plotting function
def plot_figures(points, connections, ax, circle_center, radius, labels, edge_color, point_color, circle_color):
    if circle_center:
        # Circle parameters
        theta = np.linspace(0, 2 * np.pi, 100)
        circlex = circle_center[0] + radius * np.cos(theta)
        circley = circle_center[1] + radius * np.sin(theta)

        # Plot circle
        ax.plot(circlex, circley, color=circle_color, linestyle="dotted")

    # Plot connections or triangulation
    if connections is not None:
        for (i, j) in connections:
            ax.plot([points[i][0], points[j][0]], [points[i][1], points[j][1]], color=edge_color)
    else:
        triangulation = Delaunay(points)
        ax.triplot(points[:, 0], points[:, 1], triangulation.simplices.copy(), color=edge_color)

    # Plot points and labels
    offsets = {
        "A": (0.0, -0.07),
        "B": (0.01, 0.01),
        "C": (-0.03, -0.03),
        "D": (0.01, 0.01),
    }
    for i, label in enumerate(labels):
        ax.plot(points[i][0], points[i][1], "o", color=point_color)
        offset = offsets[label]
        ax.text(points[i][0] + offset[0], points[i][1] + offset[1], label,
                verticalalignment="bottom", horizontalalignment="right", color=point_color)

    ax.axis("off")

# Define color scheme
gnbu_colormap = plt.cm.GnBu
edge_color = gnbu_colormap(0.7)
point_color = gnbu_colormap(0.5)
circle_color = gnbu_colormap(0.2)

# Define the center and radius of the circle
circle_center = [0.1844162094291, -0.209019841145]
radius = 0.6428298065757
labels = ["A", "B", "C", "D"]

# Define points and connections
points1 = np.array([[0, -0.824283289736], [0, 0.406798907447], [-0.3887200654202, -0.5001289191388], [0.8951485095589, 0.3993251586546]])
connections1 = None  # Use Delaunay triangulation

points2 = np.array([[0, -0.824283289736], [0, 0.406798907447], [-0.3887200654202, -0.5001289191388], [0.5761089674071, 0.2437780350884]])
connections2 = [(0, 1), (0, 2), (0, 3), (2, 1), (1, 3)]  # Explicit connections

points3 = np.array([[0, -0.824283289736], [0, 0.406798907447], [-0.3887200654202, -0.5001289191388], [0.5761089674071, 0.2437780350884]])
connections3 = None

# Create figure and axes
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6), dpi=300)

# Plot on each subplot
plot_figures(points1, connections1, ax1, circle_center, radius, labels, edge_color, point_color, circle_color)
plot_figures(points2, connections2, ax2, circle_center, radius, labels, edge_color, point_color, circle_color)
plot_figures(points3, connections3, ax3, None, radius, labels, edge_color, point_color, circle_color)

plt.show()
```

```python
# Define colors from the GnBu colormap
gnbu_colormap = plt.colormaps['GnBu']
edge_color = gnbu_colormap(0.7)
point_color = gnbu_colormap(0.5)
```

```python
# Use the same seed for reproducibility
np.random.seed(125)

# Define the number of points
num_points = 4

# Define a large mother triangle to encompass all points
radius = 50
offset = radius
mother_triangle = np.array(
    [[-offset, -offset], [2 * offset, -offset], [0, 2 * offset]]
)


# Function to generate random points within the triangle
def generate_points_in_triangle(triangle, num_points):
    points = []
    for _ in range(num_points):
        barycentric = np.random.rand(3)
        barycentric /= np.sum(barycentric)
        point = (
            barycentric[0] * triangle[0]
            + barycentric[1] * triangle[1]
            + barycentric[2] * triangle[2]
        )
        points.append(point)
    return np.array(points)


# Generate points within the mother triangle
points = generate_points_in_triangle(mother_triangle, num_points)

# Combine the points with the vertices of the mother triangle
all_points = np.vstack([mother_triangle, points])

# Perform incremental construction of the Delaunay triangulation
fig, axs = plt.subplots(2, 3, figsize=(15, 6), dpi=300)

# First plot: show the mother triangle and all points without connecting them
axs[0, 0].triplot(
    mother_triangle[:, 0], mother_triangle[:, 1], [[0, 1, 2]], color=edge_color
)
axs[0, 0].plot(points[:, 0], points[:, 1], "o", color=point_color)
axs[0, 0].set_aspect("equal")
axs[0, 0].set_title("Step 1: Mother triangle and $S \subset \mathbb{R}^2$")
axs[0, 0].axis("off")

# Plotting the incremental triangulation steps, starting from the second plot
for i, ax in enumerate(axs.flatten()[1:], start=1):
    selected_points = all_points[: 3 + i, :]
    tri = Delaunay(selected_points)
    title = (
        f"Step {i + 1}: After inserting {i} points"
        if i <= num_points
        else "Removal of mother triangle and DT of $S$"
    )

    if i > num_points:
        filtered_simplices = [
            s for s in tri.simplices if not any(v < 3 for v in s)
        ]
        ax.triplot(
            selected_points[:, 0],
            selected_points[:, 1],
            filtered_simplices,
            color=edge_color,
        )
    else:
        ax.triplot(
            selected_points[:, 0],
            selected_points[:, 1],
            tri.simplices,
            color=edge_color,
        )

    ax.plot(points[:i, 0], points[:i, 1], "o", color=point_color)
    ax.set_aspect("equal")
    ax.set_title(title)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
